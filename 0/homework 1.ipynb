```python
file = open('ru.txt')
words = file.readlines()
```

```python
def count_symbols(word):
    word_letters = {}
    for letter in word:
        if letter in word_letters:
            word_letters[letter] += 1;
        else:
            word_letters[letter] = 1
    return word_letters
```

```python
def contains(container, contained):
    for letter, counter in contained.items():
        if letter in container:
            if container[letter] >= counter:
                continue
            else:
                return False;
        else:
            return False;
    return True;
```

```python
suitable_words = []
initword_letters = count_symbols("ростелеком")
for word in words:
    word_letters = count_symbols(word.rstrip())
    print(word_letters)
    if contains(initword_letters, word_letters):
        suitable_words.append(word)
```
